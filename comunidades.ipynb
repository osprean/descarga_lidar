```python
import geopandas as gpd
import pandas as pd

nombre = 'EXTREMADURA'
provincias = ['CACERES', 'BADAJOZ']

comunidad = pd.concat([gpd.read_file(f'FINALES/{x}') for x in provincias])
```

```python
comunidad.drop_duplicates('geometry', inplace=True)
```

```python
comunidad
```

```python
comunidad.to_file(f'FINALES/{nombre}')
```
