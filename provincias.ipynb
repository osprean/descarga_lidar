```python
import geopandas as gpd
import pandas as pd

nombre = 'BADAJOZ'

provincia_25830 = gpd.read_file(f'FINALES/{nombre}_25830/{nombre}_25830.shp')
provincia_25830['ORIG_SRID'] = 25830
provincia_25829 = gpd.read_file(f'FINALES/{nombre}_25829/{nombre}_25829.shp')
provincia_25829['ORIG_SRID'] = 25829

provincia = pd.concat([provincia_25830, provincia_25829.to_crs(25830)])
```

```python
provincia.drop_duplicates('geometry', inplace=True)
provincia.rename(columns={'num': 'ID', 'nombre': 'NOMBRE'}, inplace=True)
```

```python
provincia.to_file(f'FINALES/{nombre}')
```
